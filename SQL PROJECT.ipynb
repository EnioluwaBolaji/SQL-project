Stay focused as we dive into building these segments and make use of the powerful CASE statement!**

**●	In this exercise, your goal is to create the segment field that buckets an athlete into one of three segments:**

**●	Tall Male, which represents a male that is at least 190 centimeters tall.**

**QUESTION 8**

**BMI bucket by sport**

**Let's examine how BMI varies across different summer sports! To gain insights, we'll construct a report that includes the following information: sport (the name of the summer sport), bmi_bucket (grouping BMI into three categories: <.25, .25-.30, >.30), and athletes (the count of unique athletes). We'll pull data from the summer_games and athletes tables. Using a CASE statement, we'll split the bmi_bucket without utilizing AND or ELSE conditions. Grouping the data by non-aggregated fields, we'll organize the report by sport and then athletes in descending order. Get ready to explore the relationship between BMI and summer sports! Let's proceed with building the query to generate the insightful report.**

**●	You are looking to understand how BMI differs by each summer sport. To answer this, set up a report that contains the following:**

**●	sport, which is the name of the summer sport**

**●	bmi_bucket, which splits up BMI into three groups: <.25, .25-.30, >.30**

**●	athletes, or the unique number of athletes**

**●	Definition: BMI = 100 * weight / (height squared).**

**●	Build a query that pulls from summer_games and athletes to show sport, bmi_bucket, and athletes.**

**●	Without using AND or ELSE, set up a CASE statement that splits bmi_bucket into three groups: '<.25', '.25-.30', and '>.30'.**

**●	Group by the non-aggregated fields.**

**●	Order the report by sport and then athletes in descending order.**


```python
%%sql
SELECT sport,
      CASE 
          WHEN 100*weight/(height^2) <.25 THEN '<.25'
          WHEN 100*weight/(height^2) <=.30 THEN '.25-.30'
          WHEN 100*weight/(height^2) >.30 THEN '>.30'  
          END AS bmi_buckets,
       COUNT(DISTINCT athlete_id) AS athletes      
FROM summer_games  AS su      
JOIN athletes  AS at 
ON su.athlete_id = at.id
GROUP BY sport, bmi_buckets
ORDER BY sport, athletes DESC
            
```

```python
%%sql
SELECT 
      region, 
    REPLACE(region,'&','and') AS character_swap,
    REPLACE(region,'.','') AS character_remove
FROM countries
GROUP BY region;
```

```python
%%sql
SELECT 
     region, 
    country_id, 
    SUM(gold) AS total_golds
FROM summer_games AS su
JOIN countries AS co
ON su.country_id = co.id
GROUP BY region,country_id
LIMIT 10
```

```python
%%sql
SELECT 
    region,
    AVG(total_golds) AS avg_total_golds
FROM
    (SELECT 
      region, 
      country_id, 
      SUM(gold) AS total_golds
FROM summer_games AS su
JOIN countries AS co
  ON su.country_id = co.id
  GROUP BY region, country_id
) AS subquery
GROUP BY region
ORDER BY avg_total_golds DESC;
```

**QUESTION 15
●	Create a report that makes use of RANK(), LEAD(), LAG(), ROLLUP(), and CUBE**


**RANK()**

```python
%%sql
WITH country_gdp AS(
SELECT 
    DISTINCT country_id,
    year,
    gdp,
    AVG(gdp) OVER (PARTITION BY year)AS country_avg_gdp
FROM country_stats

ORDER BY country_id,country_avg_gdp DESC 
)    

SELECT *,
     RANK() OVER (PARTITION BY country_id ORDER BY country_avg_gdp DESC)
FROM 
     country_gdp
LIMIT 10
```

**LEAD() , LAG()**

```python
%%sql
SELECT
      id,
     gender,
     weight,   
     age,
    LAG(age)  OVER (PARTITION BY gender  ORDER BY weight,height) AS previous_age,
    LEAD(age) OVER (PARTITION BY gender ORDER BY  weight,height) AS next_age

FROM
    athletes
ORDER BY gender,weight,height 
LIMIT 10
    
```

**ROLLUP()**

```python
%%sql
SELECT
      sport,
      year,
      SUM(bronze) AS total_bronze,
      SUM(silver) AS total_silver
FROM 
     winter_games
GROUP BY ROLLUP(sport,year) 
ORDER BY sport
```

**CUBE**

```python
%%sql
SELECT
      sport,
      year,
      SUM(bronze) AS total_bronze,
      SUM(silver) AS total_silver
FROM 
     winter_games
GROUP BY CUBE(sport,year) 
ORDER BY sport
```
